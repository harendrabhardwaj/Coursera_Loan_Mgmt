```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
```

```
dataset = pd.read_csv("loan.csv")
```

```
dataset["Loan_Status"] = dataset["Loan_Status"].replace(["N" , "Y"],[0,1])
```

```
dataset["Gender"] = dataset["Gender"].replace(["Male" , "Female"],[0,1])
```

```
dataset["Married"] = dataset["Married"].replace(["No" , "Yes"],[0,1])
```

```
dataset["Education"] = dataset["Education"].replace(["Not Graduate" , "Graduate"],[0,1])
```

```
dataset["Self_Employed"] = dataset["Self_Employed"].replace(["No" , "Yes"],[0,1])
```

```
dataset = dataset.drop(["Loan_ID"], axis = 1)
```

```
dataset = pd.get_dummies(dataset)
```

```
dataset.info()
```

```
dataset["Gender"][dataset["Gender"].isna()] = dataset["Gender"].mean()
dataset["Married"][dataset["Married"].isna()] = dataset["Married"].mean()
dataset["Self_Employed"][dataset["Self_Employed"].isna()] = dataset["Self_Employed"].mean()
dataset["LoanAmount"][dataset["LoanAmount"].isna()] = dataset["LoanAmount"].mean()
dataset["Loan_Amount_Term"][dataset["Loan_Amount_Term"].isna()] = dataset["Loan_Amount_Term"].mean()
dataset["Credit_History"][dataset["Credit_History"].isna()] = dataset["Credit_History"].mean()
```

```
dataset.info()
```

```
tm = dataset.groupby('Loan_Status').mean()
v1 = tm.iloc[0,:]
v2 = tm.iloc[1,:]
abs((v2-v1)/(v1+v2))
```

```
dataset = dataset.drop(["Self_Employed","Loan_Amount_Term"], axis = 1)
```

```
Y = dataset["Loan_Status"]
dataset.drop("Loan_Status", axis="columns", inplace=True)
X = dataset
```

# Train/Test dataset
Okay, we split our dataset into train and test set:

```
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

# Modeling(RandomForestClassifier with Scikit-learn)

```
model = RandomForestClassifier()
model.fit(X_train, y_train)
print("Accuracy of Model Using RandomForestClassifier",model.score(X_test, y_test)*100)
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
print("Loan_Status",sum(y_pred == 0))
cm = confusion_matrix(y_test,y_pred)

# jaccard
RNF_Jaccard = jaccard_similarity_score(y_test, y_pred)
print("RandomForestClassifier- jaccard accuracy = " , RNF_Jaccard)

# f1_score
RNF_f1_score = f1_score(y_test, y_pred, average='weighted') 
print("RandomForestClassifier - f1 score accuracy = " , RNF_f1_score)
```

# Modeling(Logistic Regression with Scikit-learn)

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
print("Accuracy of Model Using LogisticRegression",classifier.score(X_test, y_test)*100)
from sklearn.metrics import confusion_matrix
y_pred = classifier.predict(X_test)
print("Loan Status",sum(y_pred == 0))
cm = confusion_matrix(y_test,y_pred)

# jaccard
Log_Jaccard = jaccard_similarity_score(y_test, y_pred)
print("LogisticRegression- jaccard accuracy = " , Log_Jaccard)

# f1_score
Log_f1_score = f1_score(y_test, y_pred, average='weighted') 
print("LogisticRegression - f1 score accuracy = " , Log_f1_score)
```

# Modeling (KNN with Scikit-learn)

```
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
print("Accuracy of Model Using KNN",classifier.score(X_test, y_test)*100)
from sklearn.metrics import confusion_matrix

y_pred = classifier.predict(X_test)
print("Loan Status",sum(y_pred == 0))
cm = confusion_matrix(y_test,y_pred)

# jaccard
KNN_Jaccard = jaccard_similarity_score(y_test, y_pred)
print("KNN- jaccard accuracy = " , KNN_Jaccard)

# f1_score
KNN_f1_score = f1_score(y_test, y_pred, average='weighted') 
print("KNN - f1 score accuracy = " , KNN_f1_score)
```

# Modeling (DecisionClassifierClassifier with Scikit-learn)

```
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
print("Accuracy of Model Using DecisionTreeClassifier",classifier.score(X_test, y_test)*100)
from sklearn.metrics import confusion_matrix

y_pred = classifier.predict(X_test)
print("Loan Status",sum(y_pred == 0))
cm = confusion_matrix(y_test,y_pred)

# jaccard
DTC_Jaccard = jaccard_similarity_score(y_test, y_pred)
print("DecisionTreeClassifier- jaccard accuracy = " , DTC_Jaccard)

# f1_score
DTC_f1_score = f1_score(y_test, y_pred, average='weighted') 
print("DecisionTreeClassifier - f1 score accuracy = " , DTC_f1_score)
```

# Modeling (SVM with Scikit-learn)

```
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
print("Accuracy of Model Using SVM",classifier.score(X_test, y_test)*100)
from sklearn.metrics import confusion_matrix

y_pred = classifier.predict(X_test)
print("Loan Status",sum(y_pred == 0))
cm = confusion_matrix(y_test,y_pred)

# jaccard
SVM_Jaccard = jaccard_similarity_score(y_test, y_pred)
print("SVM- jaccard accuracy = " , SVM_Jaccard)

# f1_score
SVM_f1_score = f1_score(y_test, y_pred, average='weighted') 
print("SVM - f1 score accuracy = " , SVM_f1_score)
```
